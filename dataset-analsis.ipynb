```python
df["type"].value_counts().plot(kind='bar', title='Transaction Types', color=['skyblue', 'orange', 'lightgreen', 'salmon', 'purple', 'yellow'])
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.show()
```

```python
fraud_by_type = df.groupby('type')['isFraud'].mean().sort_values(ascending=False)
fraud_by_type.plot(kind='bar', title='Fraud Rate by Transaction Type', color=['red', 'blue', 'green', 'orange', 'purple', 'brown'])
plt.xlabel('Transaction Type')
plt.ylabel('Fraud Rate')
plt.show()
```

```python
df["amount"].describe().astype(int)
```

```python
sns.histplot(np.log1p(df["amount"] + 1), bins=100, kde=True, color='blue')
plt.title('Log-Transformed Transaction Amount Distribution')
plt.xlabel('Log(Transaction Amount + 1)')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.boxplot(data=df[df["amount"] < 50000], x='isFraud', y='amount')
plt.title('Transaction Amount by Fraud Status (Amounts < 50000)')
plt.show()
```

```python
frauds_per_step = df[df["isFraud"] == 1]["step"].value_counts().sort_index()
plt.plot(frauds_per_step.index, frauds_per_step.values, marker = 'o', color='red', label="Frauds per Step")
plt.xlabel("Step (Time)")
plt.ylabel("Number of Frauds")
plt.title("Number of Frauds Over Time")
plt.grid(True)
plt.show()
```
